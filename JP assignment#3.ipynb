```python
#Q:1 Make a calculator using python with addition,subtraction,multiplication,division and power.
num1 =  int(input("Enter the 1st number:"))
num2 =  int(input("Enter the 2nd number:"))
oprator= input("ENter the oprator +,-,x,/,** :")

if oprator == "+":
  sum = num1 + num2
  print("The sum of num1 and num2 is: ", sum)
elif oprator == "-":
  sub = num1 - num2
  print("The sum of num1 and num2 is: ", sub)
elif oprator == "x":
  multi = num1 * num2
  print("The sum of num1 and num2 is: ", multi)
elif oprator == "/":
  if num2  != 0:
    dev = num1 / num2
    print("The sum of num1 and num2 is: ", dev)
  else:
    print("Numer is not divided by 0")
elif oprator == "**":
  power = num1 ** num2
  print("The sum of num1 and num2 is: ", power)
else:
    print("Enter the correct input")
```

```python
#Q:2 write a program to check if there is any numeric value in list using for loop.
values = [1,2,3,"a","b","c"]
found = False
for i in values:
  if isinstance(i,int):
    found =True
    break
if found:
  print("Numeric value exists" )
else:
  print("No, numeric value exists")
```

```python
#Q:4 Write a python program to sum all the numeric items in a dictionary.
dic =[ {
    "name":"ali",
    "age":"10",
    "class":"8",
    "marks":34
},
        {
    "name":"aasim",
    "age":"15",
    "class":"10",
    "marks":50
},
        {
    "name":"saba",
    "age":"12",
    "class":"9",
    "marks":67
}
]
total = 0
for d in dic:
  for value in d.values():
    if isinstance(value,(int,float)):
      total += value

print("sum of numeric values is: ", total)
```

```python
#Q:5 write a program to identify duplicate values from list
num_list =[1,7,9,8,9,8,6,6,5,4]
duplicate = []
check = []
for i in num_list:
  if i in check:
    if i not in duplicate:
      duplicate.append(i)
  else:
    check.append(i)
print("The duplicate value is: ",duplicate)
```
