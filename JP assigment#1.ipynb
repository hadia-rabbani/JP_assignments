```python
# Q:7 write a program which takes 5 inputs from user for diffrent subjects's marks, total it and generate mark sheet using grades?

math = int(input("Enter the marks of Math: "))
eng = int(input("Enter the marks of Engilsh: "))
phy = int(input("Enter the marks of Physics: "))
chem = int(input("Enter the marks of Chemistry: "))
computer = int(input("Enter the marks of Computer: "))

obtained_marks = math + eng+ phy + chem + computer

percentage = (obtained_marks * 100)/500

if percentage>=80 and percentage <=100:
  grade = 'A'
elif percentage>=70 and percentage <80:
  grade = 'B'
elif percentage>=60 and percentage<70:
 grade = 'C'
elif percentage >= 50 and percentage<60:
  grade = 'D'
elif percentage>=33 and percentage<50:
  grade = 'E'
else:
  grade = 'Fail'

print(f"The Percentage is : {percentage}% so your grade is {grade} ")
```
